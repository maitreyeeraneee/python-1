```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt


# Ask user how many students they want to enter
num_students = int(input("Enter number of students: "))


# Empty lists to store names and marks
students = []
marks_list = []


# Take input from user
for i in range(num_students):
    
    name = input(f"Enter name of student {i+1}: ")
    marks = int(input(f"Enter marks of {name}: "))
    
    students.append(name)
    marks_list.append(marks)


# Convert marks list into NumPy array
marks_array = np.array(marks_list)


# Display entered data
print("\nStudent Marks:")
for i in range(num_students):
    print(students[i], ":", marks_array[i])


# Calculate average marks
average_marks = np.mean(marks_array)
print("\nAverage Marks:", average_marks)


# Highest marks
highest_marks = np.max(marks_array)
print("Highest Marks:", highest_marks)


# Lowest marks
lowest_marks = np.min(marks_array)
print("Lowest Marks:", lowest_marks)


# Find topper
topper_index = np.argmax(marks_array)

print("\nTopper:", students[topper_index])
print("Topper Marks:", marks_array[topper_index])


# Find failed students
print("\nFailed Students:")

for i in range(num_students):
    if marks_array[i] < 40:
        print(students[i], "-", marks_array[i])


# Function for grades
def assign_grade(mark):
    
    if mark >= 90:
        return "A"
    
    elif mark >= 75:
        return "B"
    
    elif mark >= 50:
        return "C"
    
    else:
        return "Fail"


# Display grades
print("\nStudent Grades:")

for i in range(num_students):
    print(students[i], ":", assign_grade(marks_array[i]))


# Plot graph
plt.bar(students, marks_array)

plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Performance Analysis")

plt.show()
```
